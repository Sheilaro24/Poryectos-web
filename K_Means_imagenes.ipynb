```python
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt

china = load_sample_image("china.jpg")
plt.imshow(china)
plt.grid(False)
plt.show()
```

```python
from sklearn.cluster import MiniBatchKMeans
import numpy as np
#Vamos a definr una variable de cuantos colores queremos que tenga la imagen final
n_colors = 64
#ahora efinimos la X
X = (china/255.0).reshape(-1,3)
#Definmos el modleo
model = MiniBatchKMeans(n_colors)
#entrenamos
labels = model.fit_predict(X)
#ahora colores
colors = model.cluster_centers_
#generaos la nueva imagen deshacemos la transformacion con la imagen destino
new_image = colors[labels].reshape(china.shape)
new_image = (255 * new_image).astype(np.uint8)
```

```python
#mostrar imagen
with plt.style.context("seaborn-white"):
  plt.figure()
  plt.imshow(china)
  plt.title("Original: 16 millones de colores")
  plt.show()
  plt.figure()
  plt.imshow(new_image)
  plt.title("Salida 64 colores")
  plt.show()
```
