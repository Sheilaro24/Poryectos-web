```python
#vmaos a usar Spicy
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#usar el matplot como si fuera seaborn
plt.style.use("seaborn")
```

```python
#cargar un dataset de prueba en este cso en sklearn tiene un dataset make_blobs que nos permite generar conjunto de puntos
from sklearn.datasets import make_blobs
X, Y = make_blobs(n_samples = 300, centers=4, random_state = 0, cluster_std=0.6)
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
#Tama√±o del k means
est = KMeans(4)
#esto es el entrenamiento
est.fit(X)
#generamos la prediccion
y_kmeans = est.predict(X)
#graficamos
plt.scatter(X[:,0],X[:,1], c=y_kmeans, s=50, cmap="rainbow")
plt.plot()
```
